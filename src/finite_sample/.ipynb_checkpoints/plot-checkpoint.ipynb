```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
path = '../../res/exp3'
def load_data(path, shuffle='no_shuffle'):
    nice = pd.read_csv(path +'/{}_irm_x_all.csv'.format(shuffle)).values
    adjust_all = pd.read_csv(path +'/{}_erm_x_all.csv'.format(shuffle)).values
    pre_treatment = pd.read_csv(path +'/{}_erm_xtat.csv'.format(shuffle)).values
    no_adjust = pd.read_csv(path +'/{}_no_adjustment_irm_x_all.csv'.format(shuffle)).values
    xt = pd.read_csv(path +'/{}_erm_xt.csv'.format(shuffle)).values
    x_and_a= pd.read_csv(path +'/{}_erm_x_and_a.csv'.format(shuffle)).values

    return nice, adjust_all, pre_treatment, no_adjust, xt, x_and_a
    
    
```

```python
def make_plot(a, b,c,d,e,f, shuffle='shuffle'):
    means = [np.mean(a), np.mean(b), np.mean(c), np.mean(d), np.mean(e), np.mean(f)]
    se= np.array([np.std(a), np.std(b), np.std(c), np.std(d), np.std(e), np.std(f)])
    barWidth = 0.7
    x=np.arange(len(means))
    y=means
    sns.set(style="white")
    sns.set_palette('Set2')
    sns.set_context('paper')
    nd = np.arange(len(means))

    plt.xticks(nd, ('(1)','(2)','(3)','(4)','(5)', '(6)'),fontsize='large')
    # plt.xlim(-0.15, 3.25)
    plt.ylim(0, 0.37)
    plt.ylabel('MAE and standard error',fontsize='large')
    fig = plt.bar(nd,y , yerr=se, width=barWidth,color=sns.color_palette("Set2",len(means)))
    plt.legend(fig, ['(1) no-adjustment','(2) adjust all','(3) adjust X and A',
                     '(4) adjust pre-treament', '(5) adjust X_t', '(6) NICE'],loc='upper right', fontsize='large',frameon=False)
    plt.tight_layout()


    plt.savefig('{}_comparisons.png'.format(shuffle),dpi=300, forma='pdf')
```

```python
nice, adjust_all, pre_treatment, no_adjust, xt, x_and_a = load_data(path, shuffle='no_shuffle')
make_plot(no_adjust, adjust_all, x_and_a, pre_treatment, xt, nice, shuffle='no_shuffle')
```
